```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
```

```python
cpu_burst = np.random.choice(np.linspace(0,1,1000000),size=(n, 1))
```

```python
alpha = 0.9
```

```python
p = [0] * n
p[0] = cpu_burst[0]
for i in range(1,n):
    p[i] = alpha*p[i-1] + (1-alpha)*cpu_burst[i-1]
    
p = np.array(p).reshape((n,1))
```

```python
plt.plot(cpu_burst)
plt.plot(p)
plt.show()
```

```python
plt.hist(cpu_burst)
plt.show()
```

```python
n = 100

cpu_burst = np.random.choice(np.linspace(0,1,1000),size=(n, 1))

errors = []
for i in np.linspace(0,1,1000):
    alpha = i
    p = [0] * n
    p[0] = cpu_burst[0]
    for i in range(1,n):
        p[i] = alpha*p[i-1] + (1-alpha)*cpu_burst[i-1]

    p = np.array(p).reshape((n,1))
    e = sum(abs(cpu_burst-p))/n
    errors.append(e)

plt.plot(np.linspace(0,1,1000), errors)
```
